```python
import numpy as np
#from scipy import misc
#misc.imread is deprecated in new version of scipy use imageio instead
import matplotlib.pyplot as plt
#from skimage import data
import imageio

photo_data = imageio.v3.imread('chinjang.jpg')
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.imshow(photo_data)
plt.title('Ass chinjang')


photo_data2 = photo_data.copy()

photo_data2[100:250,110:310 ] = 0

plt.figure(figsize=(8,8))
plt.imshow(photo_data2)
plt.title('Censor Ass chinjang')
plt.show()
```
